## 1. Convert CSV to CSVW

### 1.1 Build docker container

```python
%cd CSVToCSVW
!docker-compose build
!docker-compose up -d
print(r'http://localhost:5000')
```

### 1.2 Make api call to convert

```python
url_widget = widgets.Text(value=r'https://raw.githubusercontent.com/Mat-O-Lab/resources/main/mechanics/data/polymer_tensile/Zugversuch_eng_ETFE-Ref%2BGroesseneffekt2.TRA', description='CSV URL')
sep_widget = widgets.Text(value='auto', description='seperator')
enc_widget = widgets.Text(value='auto', description='encoding')
csv_button = widgets.Button(description='Make API call')
csv_out = widgets.Output(layout={'border': '0.5px solid black'})

@csv_out.capture()
def make_csvw(b):
    csv_out.clear_output()
    req_body = {"data_url": url_widget.value, "separator": sep_widget.value, "encoding": enc_widget.value}
    print('Making API call...')
    try:
        r = requests.post("http://localhost:5000/api", json=req_body)
        file = r.json()['filedata']
        with open('../out/csv-metadata.json', 'w') as f:
            f.write(file)
        csv_out.clear_output()
        print(file)
    except Exception as e:
        csv_out.clear_output()
        print(e)

display(url_widget, sep_widget, enc_widget, csv_button, csv_out)
csv_button.on_click(make_csvw)
```

### 1.3 Close down container

```python
!docker-compose down
%cd ..
```

## 2. MapToMethod

### 2.1 Start Docker

```python
%cd MapToMethod
!docker-compose up -d
print(r'http://localhost:5000')
```

### 2.2 Make API call

```python
jsonld_widget = widgets.Text(value=r'https://raw.githubusercontent.com/Mat-O-Lab/resources/main/mechanics/data/polymer_tensile/Zugversuch_eng_ETFE-Ref%2BGroesseneffekt2-metadata.json', description='JSON-LD', disabled=True)
method_widget = widgets.Text(value=r'https://raw.githubusercontent.com/Mat-O-Lab/MSEO/main/methods/DIN_EN_ISO_527-3.drawio.ttl', description = 'Method', disabled=True)
map_button = widgets.Button(description='Make API call')
map_out = widgets.Output(layout={'border': '0.5px solid black'})

@map_out.capture()
def make_map(b):
    map_out.clear_output()
    maplist = {
        'ExperimentDate': 'Datum', 'LenghtMeasurementInformation': 'ProbenbreiteB0', 
        'ThicknessMeasurementInformation': 'ProbendickeA0', 'ForceMeasurementInformation': 'Kraftmaximum',
        'TestTimeInterval': 'Pruefzeit'
    }
    info_lines = {
        'Datum': {'property': 'Datum', 'text': '', 'uri': jsonld_widget.value+'/Datum'}, 
        'ProbenbreiteB0':  {'property': 'Probenbreite b0', 'text': '', 'uri': jsonld_widget.value+'/ProbenbreiteB0'}, 
        'ProbendickeA0':  {'property': 'Probendicke a0', 'text': '', 'uri': jsonld_widget.value+'/ProbendickeA0'}, 
        'Kraftmaximum':  {'property': 'Kraftmaximum', 'text': '', 'uri': jsonld_widget.value+'/Kraftmaximum'}, 
        'Pruefzeit':  {'property': 'Pr\u00fcfzeit', 'text': '', 'uri': jsonld_widget.value+'/Pruefzeit'}
    }
    req_body = {'data_url': jsonld_widget.value, 'method_url': method_widget.value, 'maplist': maplist, 'info_lines': info_lines}
    print('Making API call...')
    try:
        r = requests.post("http://localhost:5000/api/mappingfile", json=req_body)
        yarrrml = json.loads(r.text)['filedata']
        with open('../out/yarrrml.yaml', 'w') as f:
            f.write(yarrrml)
        map_out.clear_output()
        print(yarrrml)
    except Exception as e:
        map_out.clear_output()
        print(e)
    
display(jsonld_widget, method_widget, map_button, map_out)
map_button.on_click(make_map)
```

### 2.3 Close down container

```python
!docker-compose down
%cd ..
```
